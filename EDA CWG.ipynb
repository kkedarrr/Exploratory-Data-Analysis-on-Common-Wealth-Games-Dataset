# Ploting bar graph of Countries and their gold medals.

```python
plt.bar(df['CGA'],df['Gold'])
plt.title('country with its gold medal',fontsize=15,color='red')
plt.xlabel('Country ',fontsize=15)
plt.ylabel('Gold medal ',fontsize=15)
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

#  Ploting bar graph of Countries and their Silver medals.

```python
plt.bar(df['CGA'],df['Silver'])
plt.title('country with its Silver medal',fontsize=15,color='red')
plt.xlabel('Country ',fontsize=15)
plt.ylabel('Silver medal ',fontsize=15)
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

#  Ploting bar graph of Countries and their Bronze medals.

```python
plt.bar(df['CGA'],df['Bronze'])
plt.title('country with its Bronze medal',fontsize=15,color='red')
plt.xlabel('Country ',fontsize=15)
plt.ylabel('Bronze medal ',fontsize=15)
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

# Ploting bar graph of countries and their respective medals

```python
df.groupby('CGA')[['Gold','Silver','Bronze']].sum().plot.bar(color=['orange','blue','red'],figsize=(18,5))
plt.ylabel('Medals')
plt.xlabel('Countries')
plt.show()
```
